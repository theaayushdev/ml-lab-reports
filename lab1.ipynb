```python
plt.figure(figsize=(8,6))
plt.scatter(data['area'], data['price'], color='red', marker=".")
plt.title('area vs price')
plt.xlabel('area ')
plt.ylabel('price ')
plt.grid(True)
plt.show()
```

```python
reg= linear_model.LinearRegression()
reg.fit(data[['area']],data['price'])
```
