```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trending.csv')


# Clean 'Video views' column: remove commas and convert to numeric
df['Video views'] = df['Video views'].str.replace(',', '').astype(float)

# Group by 'published' year and sum views
views_by_year = df.groupby('published')['Video views'].sum()


views_by_year.plot(kind='bar', color='skyblue')
plt.title("Total Video Views by Year")
plt.xlabel("Published Year")
plt.ylabel("Total Views")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
