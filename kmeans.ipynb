###  k_MEANS CLUSTERING
K-Means is an unsupervised learning algorithm used to group data into k distinct clusters based on similarity.
It tries to minimize intra-cluster variance (how close points are within the same cluster).



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[2,2,8,5,7,6,1,4]
y=[10,5,4,8,5,4,2,9]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
data=list(zip(x,y))
print(data)
```

```python
from sklearn.cluster import KMeans
data=list(zip(x,y))
inertias=[]

for i in range(1,9):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)


plt.plot(range(1,9),inertias,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(data)
plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```

```python
new_point = [[4, 4]]
cluster_label = kmeans.predict(new_point)
```

```python
print(cluster_label[0])
```

```python
# Step 4: Plot original data, clusters, centroids, and new point
plt.scatter(x, y, c=kmeans.labels_, cmap='viridis', label='Data Points')
plt.scatter(*zip(*kmeans.cluster_centers_), marker='*', color='black', s=200, label='Centroids')
plt.scatter(new_point[0][0], new_point[0][1], color='red', marker='x', s=150, label='New Point (4,4)')
plt.title("KMeans Clustering with Prediction")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
